```python
import scipy.io
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("OysterMetro_Complete_Sample_Feature_Extracted_new.csv").round(6)
df2 = pd.read_csv("OysterMetro_n12617.csv").round(6)

# Initialize an empty list to store the matching indices
matching_indices_df2 = []
print("start")
# Iterate over each row in df1
for i, row in df1.iterrows():
    # Find the index in df2 that matches the current row in df1
    try:
        matching_index = df2[(df2['EntTime'] == row['EntTime']) &  (df2['OriginLongitude'] == row['OriginLongitude']) & (df2['OriginLatitude'] == row['OriginLatitude']) & (df2['DestinationLongitude'] == row['DestinationLongitude']) & (df2['DestinationLatitude'] == row['DestinationLatitude']) & (df2['downo_sin'] == row['downo_sin']) & (df2['downo_cos'] == row['downo_cos'])].index[0]
    # Append the found index to the list
        df2.xs(matching_index)['TotalDistance']=row['TotalDistance']
    except:
        continue
```
